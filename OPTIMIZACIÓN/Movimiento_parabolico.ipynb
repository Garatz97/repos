```python
''''
En este ejercicio de ejemplo simularemos el movimiento de un proyectil.
Queremos analozar la distancia maxima que puede alcanzar el proyectil enbase al angulo de lanzamiento y la velocidad inicial.
'''
import numpy as np
import matplotlib.pyplot as plt

#Parametros del proyectil
v0 = 50 # Velocidad inicial en m/s
g = 9.81 # Aceleracion debido a la gravedad en m/s^2

#Posicion inicial del proyectil
x0 = 0.
y0 = 0.

#Función para calcular la posición del pryectil en función del tiempo
def proyectil(t,theta, y0):
    #convertir el angulo de grados a radianes
    theta_rad = np.radians(theta)

    #Calcular la posición en x e y
    x = x0 + v0 * np.cos(theta_rad) * t
    y = y0 + v0 * np.sin(theta_rad) * t - (0.5 * g * t*t)
    return x, y

# Ploteamos la trayectoria del proyectil para diferentes ángulos de lanzamiento
angles = [-45, -15, 0, 15, 30, 45, 60, 75] # Ángulos de lanzamiento en grados

plt.figure(figsize=(10, 6))
for theta in angles:
    t_max = (v0*np.sin(np.radians(theta)) + np.sqrt((v0*np.sin(np.radians(theta)))**2 + 2*g*y0)) / g
    print(t_max)
    t = np.linspace(0, t_max, num=100) # Generar 100 puntos de tiempo entre 0 y t_max
    x, y = proyectil(t, theta, y0)
    plt.plot(x, y, label=f'Ángulo:{theta}°')

plt.title('Trayectoria del Proyectil para Diferentes Ángulos de Lanzamiento')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid()
plt.show()

```

```python
'''
Ahora, podemos simular la distancia en función del ángulo de lanzamiento para encontrar el maximo alcance del proyectil.
'''
#Función para calcular el alcance del proyectil en función del ángulo de lanzamiento
#asumiendo que el proyectil aterriza cuando y=0
def distancia_maxima(theta):
    theta_rad = np.radians(theta)
    # Calcular el tiempo de vuelo total
    t_max = (v0 * np.sin(theta_rad) + np.sqrt((v0 * np.sin(theta_rad))**2 + 2 * g * y0)) / g
    # Calcular la distancia horizontal total
    x_max = v0 * np.cos(theta_rad) * t_max
    return x_max

#Evaluamos la distancia máxima para diferentes ángulos de lanzamiento

if y0 < 1e-6:
    theta_values = np.linspace(0, 90, num=91) # Ángulos de lanzamiento entre 0 y 90 grados
else:
    theta_values = np.linspace(-90, 90, num=180001) # Ángulos de lanzamiento entre -90 y 90 grados

distancias = [distancia_maxima(theta) for theta in theta_values]
# Ploteamos la distancia máxima en función del ángulo de lanzamiento
plt.figure(figsize=(10, 6))
plt.plot(theta_values, distancias)
plt.title('Distancia Máxima del Proyectil en Función del Ángulo de Lanzamiento')
plt.xlabel('Ángulo de Lanzamiento (grados)')
plt.ylabel('Distancia Máxima (m)')
plt.grid()
plt.show()
```

```python
print(f'El ángulo de lanzamiento que maximiza la distancia es: {theta_values[np.argmax(distancias)]} grados')
print(f'La distancia máxima alcanzada es: {max(distancias)} metros')
```
